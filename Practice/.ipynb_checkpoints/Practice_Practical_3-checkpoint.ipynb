```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = sns.load_dataset("iris")
```

```python
iris
```

```python
iris_grp = iris.groupby(iris["species"])
iris_grp
```

```python
iris["species"].unique()
```

```python
data1 = iris_grp.get_group("setosa")
data1
```

```python
data2 = iris_grp.get_group("virginica")
```

```python
data2
```

```python

box1 = data2.boxplot()
medians = data2.median()
for i in range(len(medians)):
    box1.annotate(medians[i],(i+1,medians[i]), ha="center", va="center", color="red", size=8 )
```
