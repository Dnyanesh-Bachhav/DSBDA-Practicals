```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
sns.histplot(df["fare"])
```

```python
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
df["fare"].value_counts()
```

```python
df["sex"].value_counts()
```

```python
count = sns.countplot(x=df["sex"])
for i in count.containers:
    count.bar_label(i)
```

```python
sns.countplot(x=df["pclass"],hue=df["sex"])
```

```python
sns.countplot( x=df["pclass"], hue=df["survived"], palette="Accent")
```

```python
plt.pie(df["sex"].value_counts(), labels=["male","female"])
plt.show()
```

```python
iris = sns.load_dataset("iris")
```

```python
iris_grp = iris.groupby(iris["species"])
```

```python
setosa = iris_grp.get_group("setosa")
setosa
```

```python
sns.kdeplot(setosa["sepal_length"], shade=True)
plt.show()
```

```python
sns.histplot(setosa["sepal_length"], kde=True)
```
